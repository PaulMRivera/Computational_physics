```python
import numpy as np
import scipy.linalg as la
from numpy.linalg import matrix_power as mpow
import matplotlib.pyplot as plt
from scipy.integrate import odeint, solve_ivp
from mpl_toolkits import mplot3d
from scipy import integrate, linalg
import pylab as plt

%matplotlib inline
```

```python
def Valores( M ):
    '''
    Parámetros
    ----------
    M : Matriz a la que se le va sacar los diferentes valores
    
    Regresa
    -------
    Los Eigenvalores
    EL Delta
    La Traza
    El determinante
    
    '''
    eigvals, eigvecs = la.eig(M)
    tr = np.trace(M)
    det = la.det(M)
    Delta = (tr)**2 - (4*det)
    print("Eigenvectores", eigvecs)
    print()
    print('Eigenvalores:', eigvals)
    print()
    print("Traza de A1:", tr) 
    print()
    print("Determinante de A1:", det)
    print()
    print("Delta de A1:", Delta)
    print()
```

```python
# SOLUCION DE LA ECUACION DIFERENCIAL 1

# DEFINIMOS UN VECTOR X: X[0]=x, X[1]=y

def dX_dt(X, t=1):
    return np.array([ X[1] , - X[0]])

# PUNTOS CRITICOS DEL SISTEMA
X_f0 = np.array([     0. ,     0.])
all(dX_dt(X_f0) == np.zeros(2) )


# SOLUCION NUMERICA PARA EL PROBLEMA 1

t = np.linspace( 0, 2*np.pi, 51 )
X0 = np.array([1, 1])                     # Condiciones iniciales
X, infodict = integrate.odeint(dX_dt, X0, t, full_output=True)
infodict['message']

x, y = X.T

f1 = plt.figure();
plt.figure(figsize=(10, 5)); 
plt.plot(t, x, 'r-', label='x(t)');
plt.plot(t, y  , 'b-', label='y(t)');
plt.grid();
plt.legend(loc='best');
plt.xlabel('tiempo');
plt.ylabel('Soluciones');
plt.title('Soluciones x(t), y(t)');
plt.show();
```

```python
# SOLUCION EN EL PLANO FASE

plt.plot(x, y, 'brown')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Solución X(t) en el plano fase')
plt.show()
```

```python
# SOLUCION DE LA ECUACION DIFERENCIAL 4

# DEFINIMOS CONDICIONES 

t = np.linspace( -5, 50, 1000 )
 
c1 = np.linspace( -0.5 ,    4 , 5 )
c2 = np.linspace(    5 , -0.2 , 5 )
c3 = np.linspace(  -10 ,   11 , 5 )

# FUNCION DE LA SOLUCION 4

def F4( t,c1, c2, c3 ):
  r=[c1*( np.exp(-2*t)), c2*np.cos( 2*t ) + c3*np.sin( 2*t )  ,
     -c2*np.sin( 2*t ) + c3*np.cos( 2*t )  ]
  return r

# GRAFICA DE LA SOLUCION 4
plt.figure(figsize = ( 12, 10 ))
ax = plt.axes( projection= "3d" )

for c,d,e in zip(c1,c2,c3):
  v = F4(t,c,d,e)
  s = np.array(v)
  ax.plot3D(s[0,:],s[1,:],s[2,:])
 
plt.xlabel("x(t)")
plt.ylabel("y(t)")
ax.set_zlabel("z(t)")
plt.title("SOLUCION DEL SISTEMA")
plt.grid()
plt.show()
```

```python
# SE DEFINE LA FUNCION DERIVADA
def f(y, t, k1, k2, b1, b2):
  x1, x2, v1, v2 = y
  dxdt=[v1, v2, -(k1+k2)*x1+k2*x2-b1*v1+k1-k2, k2*x1-k2*x2-b2*v2+k2]
  return dxdt
k1=6
k2=4
b1 = b2 = 0
y0=[1,4,0,0]

# OBTENEMOS LOS EIGEN VALORES
E=np.array([[0,0,1,0],[0,0,0,1],[-(k1+k2),k2,-b1,0],[k2,-k2,0,b2]])
eigenval, eigenvec = la.eig(E)

#DEFINIMOS LAS CONDICIONES A UTILIZAR
t=np.linspace(0,50,1000)
k1=6
k2=4
b1 = b2 = 0
y0=[1,4,0,0]
y=odeint(f,y0,t,args=(k1, k2, b1, b2))

# GRAFICAMOS EL ESPACIO FASE
plt.figure(figsize=(6,6))
plt.plot(y[0:,0],y[0:,1],color='brown')
plt.title('Espacio fase de la solución $x_2$ contra $x_1$')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.show()
```

```python
# GRAFICAMOS LA SOLUCION EN T
plt.figure(figsize=(13,9))
plt.plot(t,y[0:,0],color='brown',label='$x_1(t)$')
plt.plot(t,y[0:,1],color='blue',label='$x_2(t)$')
plt.title('Gráfico de $x_2(t)$ y $x_1(t)$ contra $t$')
plt.xlabel('$t$')
plt.ylabel('$y$')
plt.legend()
plt.show()
```
